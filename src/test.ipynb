```python
%matplotlib inline
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
data2 = pd.read_csv('results10/scaled_mean.csv')
infected = data2['scaled_mean'].to_numpy()
p1 = data2['p1'].to_numpy()
p3 = data2['p3'].to_numpy()

contour_plot2 = infected[p1==1]
for p in np.flip(np.unique(p1))[1:]:
    contour_plot2 = np.vstack((contour_plot2, infected[p1==p]))
```

```python
# plt.figure(figsize=(10,10))
plt.imshow(contour_plot2, cmap='magma', extent=[0, 1, 0, 1])
plt.xlabel('p3')
plt.ylabel('p1')
plt.colorbar()
plt.show()
```
