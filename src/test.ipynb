```python
%matplotlib inline
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from scipy.stats import sem
```

```python
data2 = pd.read_csv('scaled_mean.csv')
infected = data2['scaled_mean'].to_numpy()
p1 = data2['p1'].to_numpy()
p3 = data2['p3'].to_numpy()

contour_plot2 = infected[p1==1]
for p in np.flip(np.unique(p1))[1:]:
    contour_plot2 = np.vstack((contour_plot2, infected[p1==p]))
```

```python
plt.contourf(contour_plot2, cmap='magma', extent=[0, 1, 0, 1], origin='upper')
plt.xlabel('p3')
plt.ylabel('p1')
plt.colorbar()
plt.show()
```
